```python
import pandas as pd
train_ds = pd.read_csv(r'C:\Users\Madhan Adavani\Desktop\Loan prediction\train_ds.csv')
test_ds = pd.read_csv(r'C:\Users\Madhan Adavani\Desktop\Loan prediction\test_ds.csv')
train_ds.head()
```

```python
from pycaret.classification import *
clf1 = setup(train_ds,target='Loan_Status',ignore_features=['Loan_ID'])
```

```python
compare_models()
```

```python
tuned_ridge = tune_model('ridge')
```

```python
final_model = finalize_model(tuned_ridge)
```

```python
predictions = predict_model(final_model,data=test_ds)
```

```python
predictions.head(10)
```

```python
import numpy as np

loan_id = test_ds['Loan_ID']
pred = predictions['Label']
pred_2 = np.where(pred==1,'Y','N')
df = {'Loan_Status' : pd.Series(pred_2, index = loan_id)}
submissions = pd.DataFrame(df)
submissions.head(10)
```
